### 2nd column belongs_to_collection

```python
data.belongs_to_collection
```

```python
data.belongs_to_collection[2]
```

```python
data.belongs_to_collection.isnull().sum()
```

 - data type of belongs_to_collection is string format dictionary
 problems  need tobe solved - 
 1. convert the string to dictionary (its original form)
 2. extract 'name' from the dictionary
 3. replace NaN value with No Collection (since we have 40972)
 4. replace entire belongs column with collection column

```python
import ast
collections=[]
for i in data.belongs_to_collection:
    if i==i:
        collections.append(ast.literal_eval(i)['name'])
    else:
        collections.append("No collection")
        
```

```python
collections
```

```python
data['Collections']=collections
```

```python
data=data.drop(['belongs_to_collection'],axis=1)
```
